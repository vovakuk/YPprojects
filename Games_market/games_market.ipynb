# Введение #
Мы работаем в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Перед нами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и мы планируем кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируем ли мы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.

## Замена пропусков  
Больше всего пропусков в оценках критиков и игроков и рейтинге. Замена можеть дать ложное распределение. Пока предлагается оставить это без изменений. Посмотрим на другие пропуски.

```python
df.loc[df['name'].isnull() == True]
```

2 игры 1993го года, о которых мало что известно предлагается выбросить.

```python
df = df[pd.notnull(df['name'])]
df.loc[df['year_of_release'].isnull() == True]
```

```python
df.loc[df['name'] == 'FIFA Soccer 2004']
```

Названий игр существенно меньше количества стро из-за того, что одна игра может выходить на нескольких платформах. Это поможет заполнить пропуски. Напишем функцию, котора по известным данным релиза на другой платформе заполнит пропуски.

```python
df_years = df[pd.notnull(df['year_of_release'])]
def fill_year(row):
    rows= df_years[df_years['name'] == row['name']]
    if len(rows) > 0:
        return rows.iloc[0]['year_of_release']
    return row['year_of_release']
```

```python
df['year_of_release'] = df.apply(fill_year, axis=1)
df['year_of_release'].isna().sum()
```

146 игр остались без года издания, благо это немного. Заменим tbd в оценках игроков на nan.

```python
df.loc[df['user_score'] == 'tbd', 'user_score'] = np.nan
df['user_score'].unique()
```

# Исследовательский анализ данных

```python
year_games = (df.pivot_table(index='year_of_release', values = 'name', aggfunc = 'count')
              .sort_values('year_of_release', ascending = True))
year_games.plot(figsize=(12, 6), legend=False, title='Выпуск игр в разные годы').set(xlabel='Год релиза', ylabel='Количество')
plt.show()
```

Количество выпускаемых игр стало стремительно увеличиваться в середине 90-х годов. Возможно это связано с распространением ПК и игровых приставок среди пользователей. Посмотрим продажи по платформам. Спад продаж в последнее время связан с развитием мобильного гейминга.

```python
platform_games = (df.pivot_table(index='platform', values = 'total_sales', aggfunc = 'sum')
              .sort_values('total_sales', ascending = False))
platform_games.plot(figsize=(12, 6), kind = 'bar', legend=False, title='Выпуск игр на разных платформах').set(xlabel='Платформа', ylabel='Количество')
plt.show()
```

Наибольшие продажи у PS2. Скорее всего это связано с наиболее продолжительным периодом продаж как платформ, так и игр. Приставка PS2 продавалась 13 лет, а Х360 11 лет. Посмотрим первые 6 наболее популярных платформ подробно.

```python
top_platforms = platform_games.head(6)
top_platforms
```

```python
for i in top_platforms.index:
    pl_years=df.query('platform == @i').pivot_table(index='year_of_release', values = 'total_sales', aggfunc = 'sum').sort_values('year_of_release', ascending=False)
    pl_years['total_sales'].plot(figsize=(12, 6), legend=False,label=i, title='Выпуск игр в разные годы').set(xlabel='Год релиза', ylabel='Количество')
    plt.legend()
plt.show()
```

Видно, что в целом период активных продаж занимает примерно 10 лет. За 5 лет продажи достигают пика, а потом так же резко идут на спад. Расчитаем среднее время существования платформы, считая активными продажи свыше 5 млн копий в год.


```python
years=[]
platforms=[]
for i in platform_games.index:
    pl_years=df.query('platform == @i').pivot_table(index='year_of_release', values = 'total_sales', aggfunc = 'sum').sort_values('year_of_release', ascending=False)
    years.append(len(pl_years[pl_years['total_sales'] > 5]))
    platforms.append(i)
platform_active = pd.DataFrame(index = platforms, data=years)
platform_active.columns=['years']
platform_active.sort_values(by='years', ascending = False)
```

Много старых и неизвестных платформ, игры на которые не продавались таким тиражом. Самое длинное "время жизни" у персональных компьютеров. Это понятно - начинка в них меняется, а ПК он и остался ПК. Возьмем наши топ-6 платформ.

```python
years=[]
platforms=[]
for i in top_platforms.index:
    pl_years=df.query('platform == @i').pivot_table(index='year_of_release', values = 'total_sales', aggfunc = 'sum').sort_values('year_of_release', ascending=False)
    years.append(len(pl_years[pl_years['total_sales'] >5]))
    platforms.append(i)
platform_active = pd.DataFrame(index = platforms, data=years)
platform_active.columns=['years']
platform_active.sort_values(by='years', ascending = False)
```

## Влияние рейтинга ESRB  
Какие бывают ограничения от ESRB:
* E - для всех
* Т - старше 13
* М - младше 17
* Е10+ - старше 10
* К-А - совместно с родителями
* AO - старше 17
* ЕС - для маленьких детей

# Проверка статичтических гипотез  
Проверим гипотезу:
* Нулевая гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
* Альтернативная гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC разные  

Вычислим дисперсии для установления параметра equal_var. ddof поставим равным нулю, будем считать, что у нас полная выборка. В соответствии с небольшой длиной выборок ставим уровень значимости 0.05.

```python
pc = actual_years.query('platform == "PC"')
xone =  actual_years.query('platform == "XOne"')
print('Размер первой выборки',len(pc))
print('Размер второй выборки', len(xone))
print('Дисперсия продаж для PC {:.2f}'.format(np.var(pc['total_sales'], ddof=0)))
print('Дисперсия продаж для XOne {:.2f}'.format(np.var(xone['total_sales'],ddof=0)))
alpha = 0.05
results = st.ttest_ind(pc['total_sales'],
    xone['total_sales'],equal_var= False)

print('p-значение:', results.pvalue)

if (results.pvalue<alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод**  
Средние пользовательские рейтинги платформ Xbox One и PC разные
